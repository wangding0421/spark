### Script for plotting yearly plots 

```python
def YearlyPlots(T,ttl='',yl='',xl='',y=None,x=None,size=(10,7)):
    yearday=[i for i in range(1,366)]
    fig=figure(1,figsize=size,dpi=300)
    if shape(T)[0] != 365:
        raise ValueError("First dimension of T should be 365. Shape(T)="+str(shape(T)))
    plot_date(yearday,T, '-');
    # rotate and align the tick labels so they look better
    #fig.autofmt_xdate()
    ylabel(yl)
    xlabel(xl)
    if y is not None:
        ylim(y)
    if x is not None:
        xlim(x)
    grid()
    title(ttl)
```

### Plot the following 3 plots for each measurement:

1. A histogram from the sample values (from SortedVals) restricted between low1000 and high1000
2. Plot of mean and mean $\pm$ std
3. Number of measurements recorded each day

```python
figure(figsize=(15,30))
offset=1
for meas in STAT.keys():
    subplot(6,3,offset)
    offset+=1
    ## Your code for Histogram
    npa_RawSortedVals = STAT[meas]['SortedVals']
    npa_SortedVals = npa_RawSortedVals[where(logical_and(npa_RawSortedVals>=STAT[meas]['low100'],\
                                                         npa_RawSortedVals<=STAT[meas]['high100']))[0]]
    hist(npa_SortedVals, bins=100)
    title('%s restricted Hist'%(meas))
    
    subplot(6,3,offset)
    offset+=1
    ## Your code for mean and mean +- std
    YearlyPlots(STAT[meas]['Mean'] + sqrt(STAT[meas]['Var']))
    YearlyPlots(STAT[meas]['Mean'])
    YearlyPlots(STAT[meas]['Mean'] - sqrt(STAT[meas]['Var']))
    title('%s mean +- std'%(meas))
    
    subplot(6,3,offset)
    offset+=1
    ## Your code for number of measurements
    YearlyPlots(STAT[meas]['NE'],ttl='%s Counts'%(meas))    
```

### Plot the following 3 plots for each measurement:

1. The percentage of variance explained by top-k eigen vectors for k between 1 to 9
2. Plot of mean and mean $\pm$ std
3. Plot of top 3 eigenvectors

```python
figure(figsize=(15,30))
offset=1
npa_k = np.arange(10)
for meas in STAT.keys():
    subplot(6,3,offset)
    offset+=1
    ## Your code for percentage of variance explained
    npa_EigVal = STAT[meas]['eigval']
    f_SumVar = sum(npa_EigVal)
    npa_PerVarExplained = zeros(len(npa_k))
    for i in npa_k:
        npa_PerVarExplained[i] = sum(npa_EigVal[0:i])/f_SumVar
    plot(npa_k, npa_PerVarExplained)
    
    subplot(6,3,offset)
    offset+=1
    ## Your code for mean and mean +- std
    YearlyPlots(STAT[meas]['Mean'] + sqrt(STAT[meas]['Var']))
    YearlyPlots(STAT[meas]['Mean'])
    YearlyPlots(STAT[meas]['Mean'] - sqrt(STAT[meas]['Var']))
    title('%s mean +- std'%(meas))    
    
    subplot(6,3,offset)
    offset+=1
    ## Your code for top-3 eigenvectors
    YearlyPlots(STAT[meas]['eigvec'].T[0:3].T)
    title('%s Top 3 EigenVecs'%(meas))      
```
